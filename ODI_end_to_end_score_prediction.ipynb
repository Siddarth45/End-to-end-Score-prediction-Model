```python
consistent_teams=['Australia', 'New Zealand', 'England', 'South Africa',
       'West Indies', 'Sri Lanka', 'Pakistan', 'India','Afghanistan',
       'Bangladesh', 'Zimbabwe', 'Ireland']
```

```python
df=df[df['bowling_team'].isin(consistent_teams)]
```

```python
df=df[df['batting_team'].isin(consistent_teams)]
```

```python
df=df.sort_values(['match_id', 'innings'], ascending=[ True,True])
```

```python
df['overs']=df['ball']
```

```python
df.head(1)
```

```python
df['ball']=df['ball'].astype(str)
```

```python
df['over']=df['ball'].str[0:-1]
```

```python
df['over']=df['over'].str.replace('.','')
```

```python
df['ball']=df['ball'].str[-1:]
```

```python
df.head(1)
```

```python
df['total_runs']=df['extras']+df['runs_off_bat']
```

```python
df.head(5)
```

```python
df.columns.unique()
```

```python
df=df[['match_id', 'innings','batting_team', 'bowling_team','overs', 'over', 'ball','total_runs','player_dismissed' ]]
```

```python
df=df.rename(columns={"match_id":"id"})
```

```python
df
```

```python
df=df.replace(np.nan,0)
```

```python
df
```

```python
df['total']=df.groupby(['id','innings'])['total_runs'].transform('sum')
```

```python
df['total_score']=df.groupby(['id','innings'])['total_runs'].apply(lambda x:x.cumsum())
```

```python
df.head(7)
```

```python
tmp1=df.groupby(['id','innings'])['total_runs'].rolling(min_periods=1, window=60).sum().reset_index()
```

```python
tmp1['total_runs']
```

```python
df['prev_60balls_run']=tmp1['total_runs'].to_list()
```

```python
df
```

```python
df['player_dismissed']=np.where(df['player_dismissed']== 0,0,1)
tmp2=df.groupby(['id','innings'])['player_dismissed'].rolling(min_periods=1, window=60).sum().reset_index()
df['prev_60balls_wickets']=tmp2['player_dismissed'].to_list()
```

```python
df['total_wickets']=df.groupby(['id','innings'], group_keys=False)['player_dismissed'].apply(lambda x:x.cumsum())
```

```python
df['prev_60balls_dots']=df['total_runs']
df['prev_60balls_dots']=np.where(df['prev_60balls_dots']== 0,1,0)
df.groupby(['id','innings'])['prev_60balls_dots'].rolling(min_periods=1, window=60).sum().reset_index()
```

```python
tmp3=df.groupby(['id','innings'])['prev_60balls_dots'].rolling(min_periods=1, window=60).sum().reset_index()
df['prev_60balls_dots']=tmp3['prev_60balls_dots'].to_list()
```

```python
df.head(10)
```

```python
df['prev_60balls_boundaries']=df['total_runs']
df['prev_60balls_boundaries']=np.where(df['prev_60balls_boundaries']>3,1,0)
df.groupby(['id','innings'])['prev_60balls_boundaries'].rolling(min_periods=1, window=60).sum().reset_index()
```

```python
tmp4=df.groupby(['id','innings'])['prev_60balls_boundaries'].rolling(min_periods=1, window=60).sum().reset_index()
df['prev_60balls_boundaries']=tmp4['prev_60balls_boundaries'].to_list()
```

```python
df.head(60)
```

```python
convert_dict={'prev_60balls_run':int,	'prev_60balls_wickets':int,	'prev_60balls_dots':int,'prev_60balls_boundaries':int}
df=df.astype(convert_dict)
```

# ML

```python
x = df.drop(labels = ['total'], axis = 1)
y = df['total'].values
```

```python
x_df = pd.DataFrame(x, columns=['batting_team', 'bowling_team', 'overs',
       'total_score','total_wickets',
       'prev_60balls_run', 'prev_60balls_wickets',
       'prev_60balls_dots', 'prev_60balls_boundaries'])
y_df = pd.DataFrame(y, columns=['total'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size = 0.25, random_state = 42, stratify = y_df)
```

```python
x_train
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('ODI_pipe.pkl','wb'))
```
